```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')
```

```python
from google.colab import drive
drive.mount("/content/drive")
```

```python
data = pd.read_csv("/content/drive/MyDrive/ToBigs_session/Week2/과제/assignment3_creditcard.csv")
```

```python
data.head()
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
y.value_counts()
```

```python
X.describe()
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support

## performance metrics
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
precision, recall, fscore, support = \
    precision_recall_fscore_support(y_test, y_pred)

print('Accuracy   : %.3f' %accuracy)
print('Precision  : %.3f' %precision[0])
print('Recall     : %.3f' %recall[0])
print('Specificyty: %.3f' %recall[1])
print('F1-Score   : %.3f' %fscore[0])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10, stratify = y)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support

## performance metrics
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
precision, recall, fscore, support = \
    precision_recall_fscore_support(y_test, y_pred)

print('Accuracy   : %.3f' %accuracy)
print('Precision  : %.3f' %precision[0])
print('Recall     : %.3f' %recall[0])
print('Specificyty: %.3f' %recall[1])
print('F1-Score   : %.3f' %fscore[0])
```
