```python
data.describe()
```

```python
def outliers_IQR(data):
   q1=data.quantile(0.25, numeric_only=True)
   q3=data.quantile(0.75, numeric_only=True)
   IQR=q3 - q1
   outliers = data[((data < (q1 - 1.5 * IQR)) | (data > (q3 + 1.5 * IQR)))]

   return outliers
```

```python
# 원하는 열 이름 바꾸면서 이상치가 있는 데이터만 남기기
column = "TotalWorkingYears"
data_col = data[[column]]
data_outliers = outliers_IQR(data_col)
data_outliers[data_outliers.notna().any(axis=1)]
```

```python
# 이상치 개수
data_outliers.notna().sum()
```
